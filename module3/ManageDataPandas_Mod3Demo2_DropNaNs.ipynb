## How to drop NaNs?

### Dropping rows: Any

The easiest method to drop null values is just the `.dropna()` method which removes *rows* if *any* columns that have a `NaN`.

```python
drinks.dropna()
```

However, notice something interesting:

```python
drinks
```

This method `.dropna()` by default doesn't do changes inplace, to do so you need to pass `inplace=True`, or set the output to another variables.

```python
df = drinks.copy()
df.dropna(inplace=True)
df
```

```python
drinks
```

Notice we made a copy, did the changes inplace and then it changed; but for the original df it was as normal

### Dropping rows: More than one

However, sometimes you want to drop the row only if more than `N` columns have `NaNs`. Let's see how to do that:

```python
drinks.dropna(thresh=len(drinks.columns)-1)    # We allow 1 NaN
```

This command enables us to keep at most 1 NaN, if we remove it notice we go back to the original dropna.

```python
drinks.dropna(thresh=len(drinks.columns))  # Same as dropping rows with any NaN
```

We can also droprows based on Nan on only some columns:

```python
drinks.dropna(subset=['beer_servings', 'spirit_servings', 'wine_servings'])
```

Notice now we are keeping the rows before we dropped

## Dropping columns

```python
drinks.dropna(axis='columns')
```

Here we are dropping columns with any Nan, therefore only dropping the continent.
